```python
import os
import uuid

def is_uuid_name(name: str) -> bool:
    """Return True if the string is a valid UUID."""
    try:
        uuid.UUID(name)
        return True
    except ValueError:
        return False


def rename_midis_to_uuid():
    base_dir = os.getcwd()
    midi_dir = os.path.join(base_dir, "MIDI")

    if not os.path.isdir(midi_dir):
        print(f'Error: "MIDI" folder not found at {midi_dir}')
        return

    renamed_count = 0
    skipped_count = 0

    for filename in os.listdir(midi_dir):
        lower = filename.lower()

        if lower.endswith(".mid") or lower.endswith(".midi"):
            old_path = os.path.join(midi_dir, filename)

            name_without_ext, ext = os.path.splitext(filename)

            # ⭐ Skip if already UUID
            if is_uuid_name(name_without_ext):
                #print(f"Skipping (already UUID): {filename}")
                skipped_count += 1
                continue

            new_name = f"{uuid.uuid4()}{ext}"
            new_path = os.path.join(midi_dir, new_name)

            os.rename(old_path, new_path)

            print(f"{filename}  ->  {new_name}")
            renamed_count += 1

    print(f"\nDone. Renamed {renamed_count} MIDI files. Skipped {skipped_count}.")

rename_midis_to_uuid()
```
